# 03 - Finalna evaluacija TF-IDF modela

Ova sveska prati skriptu `scripts/04_evaluate_tfidf_test.py`. Test skup se koristi tek nakon izbora hiperparametara, za finalnu procenu modela.

```python
from pathlib import Path

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
if not (ROOT / "data").exists():
    ROOT = ROOT.parent

ROOT
```

```python
import joblib
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix

TEST_PATH = ROOT / "data" / "processed" / "splits" / "test.csv"
MODEL_PATH = ROOT / "models" / "tfidf_logreg_pipeline.joblib"
RESULTS_PATH = ROOT / "reports" / "04_tfidf_test_results.txt"

test_df = pd.read_csv(TEST_PATH)
model = joblib.load(MODEL_PATH)

x_test = test_df["message"]
y_test = test_df["label"]

test_df.head()
```

## Predikcija i metrike na test skupu

```python
test_predictions = model.predict(x_test)

test_report = classification_report(y_test, test_predictions)
test_confusion_matrix = confusion_matrix(y_test, test_predictions, labels=["ham", "spam"])

print(test_report)
pd.DataFrame(test_confusion_matrix, index=["stvarno_ham", "stvarno_spam"], columns=["pred_ham", "pred_spam"])
```

## Cuvanje finalnog test izvestaja

```python
RESULTS_PATH.parent.mkdir(parents=True, exist_ok=True)

with open(RESULTS_PATH, "w", encoding="utf-8") as file:
    file.write("TEST SKUP\n")
    file.write(test_report)
    file.write("\n\nMATRICA KONFUZIJE NA TEST SKUPU\n")
    file.write(str(pd.DataFrame(test_confusion_matrix, index=["stvarno_ham", "stvarno_spam"], columns=["pred_ham", "pred_spam"])))

print(f"Rezultati testa su sacuvani u: {RESULTS_PATH}")
```
